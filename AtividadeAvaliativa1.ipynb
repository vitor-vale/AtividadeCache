```python
import numpy as np
import random
import math
import time
from collections import OrderedDict
from collections import defaultdict

print("Bibliotecas Inseridas")
```

```python
#Função para quebrar o livro em arquivos menores
input_file = 'LivroDanteEx.txt'
output_prefix = 'parte_'

with open(input_file, 'r') as f:
    text = f.read()

# dividir o texto em palavras
words = text.split()
total_words = len(words)

# criar um loop para dividir o texto em arquivos menores
for i in range(math.ceil(total_words/1252)):
    # nome do arquivo de saída
    output_file = f'{output_prefix}{i+1}.txt'

    # gravar as próximas 1000 palavras em um novo arquivo de texto
    start = i*1000
    end = start + 1000
    with open(output_file, 'w') as f:
        f.write(' '.join(words[start:end]))
```

```python
usu = 0
print("Insira o número do texto que você deseja ler")
print("Caso deseje encerrar o programa, digite 0")
print("Caso deseje entrar no modo simulação digite -1")
selec = int(input("Opção desejada: "))

if selec == -1:
  with open("relatorio.txt", "w") as relatorio:
    print("Entrando no modo simulação")
    for usu in range(3):
      relatorio.write("Usuário: ")
      usu = str(usu)

      relatorio.write(usu)
      relatorio.write("\n")
      class LRUCache:
        def __init__(self, capacity):
            self.capacity = capacity
            self.cache = OrderedDict()

        def get(self, key):
            if key not in self.cache:
                return None
            value = self.cache.pop(key)
            self.cache[key] = value
            return value

        def put(self, key, value):
            if key in self.cache:
                self.cache.pop(key)
            elif len(self.cache) >= self.capacity:
                self.cache.popitem(last=False)
            self.cache[key] = value

      cache = LRUCache(10)
      request_count = {'cache_hit': 0, 'cache_miss': 0}
      total_time = 0

      relatorio.write("##############LRU################## \n")

    # Solicitação de arquivos aleatórios
      relatorio.write("##############ALEATÓRIOS PUROS################## \n")
      for i in range(200):
          file_number = random.randint(1, 100)
          start_time = time.time()
          file_content = cache.get(file_number)
          end_time = time.time()
          total_time += end_time - start_time
          if file_content is None:
              request_count['cache_miss'] += 1
              file_content = f"Conteúdo do arquivo {file_number}"
              cache.put(file_number, file_content)
              relatorio.write(file_content)
              relatorio.write(", ")
              relatorio.write("file miss ")
              req = str(request_count['cache_miss'])
              relatorio.write(req)
              relatorio.write(" Tempo Necessário: ")
              tempoTot = str(total_time)
              relatorio.write(tempoTot)
              relatorio.write("\n")
          else:
              request_count['cache_hit'] += 1
              relatorio.write(file_content)
              relatorio.write(", ")
              relatorio.write("file hit ")
              req = str(request_count['cache_hit'])
              relatorio.write(req)
              relatorio.write(" Tempo Necessário: ")
              tempoTot = str(total_time)
              relatorio.write(tempoTot)
              relatorio.write("\n")

      relatorio.write("##############POISSON################## \n")

      for i in range(200):
        #poisson
        file_number = np.random.poisson(100)
        start_time = time.time()
        file_content = cache.get(file_number)
        end_time = time.time()
        if file_content is None:
            request_count['cache_miss'] += 1
            file_content = f"Conteúdo do arquivo {file_number}"
            cache.put(file_number, file_content)
            cache.put(file_number, file_content)
            relatorio.write(file_content)
            relatorio.write(", ")
            relatorio.write("file miss ")
            req = str(request_count['cache_miss'])
            relatorio.write(req)
            relatorio.write(" Tempo Necessário: ")
            tempoTot = str(end_time)
            relatorio.write(tempoTot)
            relatorio.write("\n")
        else:
            request_count['cache_hit'] += 1
            relatorio.write(file_content)
            relatorio.write(", ")
            relatorio.write("file hit ")
            req = str(request_count['cache_hit'])
            relatorio.write(req)
            relatorio.write(" Tempo Necessário: ")
            tempoTot = str(end_time)
            relatorio.write(tempoTot)
            relatorio.write("\n")
      relatorio.write("##############ENTRE 30 e 40################## \n")
      for i in range(200):
        #entre 30 e 40
        if random.random() < 0.33:
          file_number = random.randint(30, 40)
        else:
          file_number = random.randint(1, 29) if random.random() < 0.5 else random.randint(41, 100)
        start_time = time.time()
        file_content = cache.get(file_number)
        end_time = time.time()
        if file_content is None:
            request_count['cache_miss'] += 1
            file_content = f"Conteúdo do arquivo {file_number}"
            cache.put(file_number, file_content)
            relatorio.write(file_content)
            relatorio.write(", ")
            relatorio.write("file miss ")
            req = str(request_count['cache_miss'])
            relatorio.write(req)
            relatorio.write(" Tempo Necessário: ")
            tempoTot = str(end_time)
            relatorio.write(tempoTot)
            relatorio.write("\n")
        else:
            request_count['cache_hit'] += 1
            relatorio.write(file_content)
            relatorio.write(", ")
            relatorio.write("file hit ")
            req = str(request_count['cache_hit'])
            relatorio.write(req)
            relatorio.write(" Tempo Necessário: ")
            tempoTot = str(end_time)
            relatorio.write(tempoTot)
            relatorio.write("\n")

    ######LFU#####
      relatorio.write("##############LFU################## \n")

      class LFUCache:
          def __init__(self, capacity):
              self.capacity = capacity
              self.cache = {}
              self.freq = defaultdict(list)
              self.min_freq = 0

          def get(self, key):
              if key not in self.cache:
                  return None

              value, freq = self.cache[key]
              self.freq[freq].remove(key)
              if not self.freq[freq] and freq == self.min_freq:
                  self.min_freq += 1

              freq += 1
              self.freq[freq].append(key)
              self.cache[key] = (value, freq)
              return value

          def put(self, key, value):
              if self.capacity == 0:
                  return

              if key in self.cache:
                  self.cache[key] = (value, self.cache[key][1])
                  self.get(key)
              else:
                  if len(self.cache) == self.capacity:
                      evict_key = self.freq[self.min_freq].pop(0)
                      del self.cache[evict_key]

                  self.cache[key] = (value, 1)
                  self.freq[1].append(key)
                  self.min_freq = 1

      cache = LFUCache(10)

      relatorio.write("##############ALEATÓRIOS PUROS################## \n")

      for i in range(200):
            # Gerar uma chave aleatória entre 1 e 100
            key = random.randint(1, 100)
            start_time = time.time()
            value = cache.get(key)
            end_time = time.time()
            if value is not None:
                relatorio.write("Chave: ")
                chave = str(key)
                relatorio.write(chave)
                relatorio.write(", ")
                relatorio.write("file hit")
                relatorio.write(", ")
                relatorio.write(" Tempo Necessário: ")
                temp = str(end_time)
                relatorio.write(temp)
                relatorio.write("\n")
            else:
                relatorio.write("Chave: ")
                chave = str(key)
                relatorio.write(chave)
                relatorio.write(", ")
                relatorio.write("file miss")
                relatorio.write(", ")
                relatorio.write(" Tempo Necessário: ")
                temp = str(end_time)
                relatorio.write(temp)
                relatorio.write("\n")
                cache.put(key, f"valor-{key}")

      relatorio.write("##############POISSON################## \n")

      for i in range(200):
        #poisson
          key =  file_number = np.random.poisson(100)
          start_time = time.time()
          value = cache.get(key)
          end_time = time.time()
          if value is not None:
                relatorio.write("Chave: ")
                chave = str(key)
                relatorio.write(chave)
                relatorio.write(", ")
                relatorio.write("file hit")
                relatorio.write(", ")
                relatorio.write(" Tempo Necessário: ")
                temp = str(end_time)
                relatorio.write(temp)
                relatorio.write("\n")
          else:
                relatorio.write("Chave: ")
                chave = str(key)
                relatorio.write(chave)
                relatorio.write(", ")
                relatorio.write("file miss")
                relatorio.write(", ")
                relatorio.write(" Tempo Necessário: ")
                temp = str(end_time)
                relatorio.write(temp)
                relatorio.write("\n")
                cache.put(key, f"valor-{key}")

      relatorio.write("##############ENTRE 30 e 40################## \n")

      for i in range(200):
          #30 e 40
          if random.random() < 0.33:
            key = random.randint(30, 40)
          else:
            key = random.randint(1, 29) if random.random() < 0.5 else random.randint(41, 100)
          start_time = time.time()
          value = cache.get(key)
          end_time = time.time()
          if value is not None:
                relatorio.write("Chave: ")
                chave = str(key)
                relatorio.write(chave)
                relatorio.write(", ")
                relatorio.write("file hit")
                relatorio.write(", ")
                relatorio.write(" Tempo Necessário: ")
                temp = str(end_time)
                relatorio.write(temp)
                relatorio.write("\n")
          else:
                relatorio.write("Chave: ")
                chave = str(key)
                relatorio.write(chave)
                relatorio.write(", ")
                relatorio.write("file miss")
                relatorio.write(", ")
                relatorio.write(" Tempo Necessário: ")
                temp = str(end_time)
                relatorio.write(temp)
                relatorio.write("\n")
                cache.put(key, f"valor-{key}")

      ##FIFO###

      relatorio.write("##############FIFO################## \n")

      cache_size = 10
      cache = []

      def add_to_cache(file):
          if len(cache) < cache_size:
              cache.append(file)
          else:
              cache.pop(0)
              cache.append(file)

      def is_in_cache(file):
          return file in cache

      def access_file(file):
          if is_in_cache(file):
              relatorio.write("Cache Hit, ")
              filestr = str(file)
              relatorio.write(filestr)
              Tempo = time.time() - start_time
              Tempo = str(Tempo)
              relatorio.write(Tempo)
              relatorio.write("\n")
          else:
              relatorio.write("Cache Miss, ")
              filestr = str(file)
              relatorio.write(filestr)
              Tempo = time.time() - start_time
              Tempo = str(Tempo)
              relatorio.write(Tempo)
              relatorio.write("\n")
              add_to_cache(file)


      relatorio.write("##############ALEATÓRIOS PUROS################## \n")
      for i in range(200):
          file = random.randint(1, 100)
          access_file(file)

      relatorio.write("##############POISSON################## \n")
      for j in range(200):
          file = np.random.poisson(100)
          access_file(file)

      relatorio.write("##############ENTRE 30 E 40################## \n")
      for z in range(200):
        if random.random() < 0.33:
          file = random.randint(30, 40)
          access_file(file)
        else:
          file = random.randint(1, 29) if random.random() < 0.5 else random.randint(41, 100)
          access_file(file)

  print("Relatório Gerado: relatorio.txt")
elif selec == 0:
  print("Encerrando Programa")
else:
  print("Abrindo texto selecionado")
  selec_str = str(selec)
  nomeArq = "parte_"+selec_str+".txt"
  with open(nomeArq, 'r') as f:
    text = f.read()
    print(text)
```
